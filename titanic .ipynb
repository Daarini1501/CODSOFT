```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset from Seaborn (or use your own CSV file)
df = sns.load_dataset("titanic")

# Display the first 5 rows
print(" First 5 Rows of the Dataset:")
display(df.head())

# Display dataset shape (rows, columns)
print(f"\n Dataset Shape: {df.shape}")

# Check for missing values
print("\n Missing Values:")
print(df.isnull().sum())

# Display column data types
print("\n Column Data Types:")
print(df.dtypes)

# Summary statistics
print("\n Summary Statistics:")
display(df.describe())

# Count of Survived vs Not Survived
plt.figure(figsize=(6, 4))
sns.countplot(x="survived", data=df, palette="coolwarm")
plt.title("Survival Count (0 = No, 1 = Yes)")
plt.show()

# Count of passengers by class
plt.figure(figsize=(6, 4))
sns.countplot(x="class", data=df, palette="Set2")
plt.title("Passenger Class Distribution")
plt.show()

# Survival rate by gender
plt.figure(figsize=(6, 4))
sns.barplot(x="sex", y="survived", data=df, palette="magma")
plt.title("Survival Rate by Gender")
plt.show()

# Age distribution
plt.figure(figsize=(8, 4))
sns.histplot(df["age"], bins=30, kde=True, color="blue")
plt.title("Age Distribution of Passengers")
plt.show()

# Boxplot of fares by class
plt.figure(figsize=(8, 5))
sns.boxplot(x="class", y="fare", data=df, palette="cool")
plt.title("Fare Distribution by Passenger Class")
plt.show()
```
